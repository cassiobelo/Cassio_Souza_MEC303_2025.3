4. Análise de modos e frequências naturais

Descrição da Implementação — Modos e Frequências Naturais
1. Definição das matrizes do sistema
* Matriz de rigidez K:
  * Representa a resistência do sistema às deformações
  * Modelo simples com dois graus de liberdade e acoplamento entre os nós

* Matriz de massa M:
  * Considerada como identidade (massa unitária em cada grau de liberdade)

2. Resolução do problema de autovalores
* Utilizou-se a função eigh(K, M) da biblioteca scipy.linalg para resolver o problema:
K ⋅ 𝜙 = 𝜆 ⋅ M ⋅ 𝜙
* Os autovalores 𝜆 representam os quadrados das frequências naturais
* Os autovetores 𝜙 representam os modos de vibração normalizados

3. Cálculo das frequências naturais
* As frequências foram obtidas por:
f = 𝜆 ^(1/2) / 2𝜋
* O resultado é dado em **Hertz (Hz)**

4. Resultados
* O script imprime:
  * As frequências naturais dos dois modos
  * Os modos de vibração correspondentes, normalizados

```python
import numpy as np
from scipy.linalg import eigh

# Matriz de rigidez K (2 DOFs)
K = np.array([
    [1000, -1000],
    [-1000, 1000]
])

# Matriz de massa M (2 DOFs)
M = np.array([
    [1, 0],
    [0, 1]
])

# Resolver problema de autovalores generalizado
eigvals, eigvecs = eigh(K, M)

# Frequências naturais (Hz)
frequencias = np.sqrt(eigvals) / (2 * np.pi)

print("Frequências naturais (Hz):")
for i, f in enumerate(frequencias):
    print(f"Modo {i+1}: {f:.4f} Hz")

print("\nModos de vibração (normalizados):")
print(eigvecs)
```
